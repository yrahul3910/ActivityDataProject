```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df[:5]
```

```python
# Remove last few columns that are not data
df = df[pd.notnull(df['> Tag/Category'])]
```

```python
categories = df['> Tag/Category']
```

```python
categories = list(categories)
```

```python
categories[:5]
```

```python
unique = np.unique(categories)
```

```python
int_to_str = {}
str_to_int = {}

for i, cat in enumerate(unique):
    int_to_str[i] = cat
    str_to_int[cat] = i
```

```python
category_ints = [str_to_int[x] for x in categories]
```

```python
category_ints[:6]
```

```python
sequences = [category_ints[i:i+6] for i in range(len(category_ints))]
```

```python
len(sequences)
```

```python
len(category_ints)
```

```python
sequences[-6:]
```

```python
sequences = sequences[:-5]
```

```python
sequences[-5:]
```

```python
data = { tuple(x[:5]): x[5] for x in sequences }
```

```python
data
```

```python
# For future use, we will save these into a file.
import pickle
```

```python
with open('sequences.pickle', 'wb') as f:
    pickle.dump(data, f)
```
