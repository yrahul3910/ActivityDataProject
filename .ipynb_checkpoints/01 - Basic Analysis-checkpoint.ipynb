## Plot frequency of work every minute

```python
work_rows = df[(df['> Tag/Category'] == 'work') | (df['> Tag/Category'] == 'study') | 
               (df['> Tag/Category'] == 'code') | (df['> Tag/Category'] == 'blog')]
```

```python
work_rows.info()
```

```python
work_times = []

for _, row in work_rows.iterrows():
    start = row['> Start Date']
    end = row['> End Date']
    
    start_min = int(parser.parse(start).strftime('%H')) * 60 + int(parser.parse(start).strftime('%M'))
    end_min = int(parser.parse(end).strftime('%H')) * 60 + int(parser.parse(end).strftime('%M'))
    
    work_times.extend(list(range(start_min, end_min + 1)))
```

```python
len(work_times)
```

```python
fig, ax = plt.subplots();
fig.set_size_inches(8, 8);
sns.distplot(work_times, ax=ax, hist=False, kde_kws = {'shade': True});
```

So between around 6 PM and 9 PM I'm the most productive. Verifying this with just hourly plots:

```python
work_rows = df[(df['> Tag/Category'] == 'work') | (df['> Tag/Category'] == 'study') | 
               (df['> Tag/Category'] == 'code') | (df['> Tag/Category'] == 'blog')]
work_rows['> Start Date'] = [parser.parse(x).strftime('%H') for x in work_rows['> Start Date']]
```

```python
work_rows['> Start Date'].value_counts().plot(kind='bar');
```
